### Simple line control environment

We will begin with a very simple control problem consisting in maintaining a moving object along a straight line.
We can control its acceleration as shown in the following picture.
The objective is to keep minimum the distance between the object and the center line as long as possible.
To make the problem a bit interesting, we constrain the norm of the acceleration to be larger than a given value $a_{min}$: $\forall t > 0, \Vert a(t) \Vert_2 \geqslant a_{min}$.

![Line control environment](line_control_diagram.png)

#### Mathematical modeling of the Reinforcement Learning problem

We want to learn to control the object using Reinforcement Learning.
The state of the system must contain the minimal information required to update the physics of the system from the action, i.e. the acceleration of the object.
In this case, we need at least the position and the speed of the object, thus the state will be defined at any time $t$ by $s(t)=(x(t), y(t), v_x(t), v_y(t))$
Moreover, for classical RL algorithms to apply, we need to discretize the time every $\Delta t$ time units.
We can now approximate the physics of the object movement with the following equations, knowing the initial state $s(0)=(x(0), y(0), v_x(0), v_y(0))$:
- $\forall t > 0, v_x(t+\Delta t) = v_x(t) + a_x(t) \cdot \Delta t$ ;
- $\forall t > 0, x(t+\Delta t) = x(t) + v_x(t) \cdot \Delta t$ ;
- $\forall t > 0, v_y(t+\Delta t) = v_y(t) + a_y(t) \cdot \Delta t$ ;
- $\forall t > 0, y(t+\Delta t) = y(t) + v_y(t) \cdot \Delta t$ ;

Since we want to keep minimum the distance between the center line and the object, we will model the reward signal at any time $t$ by $r(t) = e^{-\vert y(t) \vert}$.
By doing so, an RL agent who will try to maximize the cumulated sum of (discounted) rewards will try to keep $y(t)$ as close as possible to $0$ at any time step.
There are two possible ways to enforce the constraint $\Vert a(t) \Vert_2 \geqslant a_{min}$ in RL:
- either by ensuring that the algorithm will only select such actions ;
- or by associating a very large penalty (i.e. negative reward) to transitions labelled with such actions.

````python
from typing import *
from numpy.typing import ArrayLike

from skdecide import *
from skdecide.builders.domain import *
from skdecide.hub.space.gym import *
import numpy as np
import pygame
from pygame import gfxdraw
from math import exp, fabs


HORIZON = 200
ACCELERATION_MIN = 0.5
PENALTY = -5.0


class D(RLDomain, UnrestrictedActions, FullyObservable, Renderable):
    T_state = ArrayLike  # Type of states
    T_observation = T_state  # Type of observations
    T_event = ArrayLike  # Type of events
    T_action = T_event
    T_value = float  # Type of transition values (rewards or costs)
    T_info = None  # Type of additional information in environment outcome


class SimpleLineControlDomain(D):
    """This class implements the deterministic line control environment."""

    metadata = {"render_modes": ["human", "rgb_array"], "render_fps": 50}

    def __init__(self):
        # Initialize action and observation spaces compatible with Gym-style algorithms.
        inf = np.finfo(np.float32).max
        self.action_space = BoxSpace(
            np.array([-10.0, -10.0], dtype=np.float32),
            np.array([10.0, 10.0], dtype=np.float32),
            dtype=np.float32,
        )
        self.observation_space = BoxSpace(
            np.array([-inf, -inf, -inf, -inf], dtype=np.float32),
            np.array([inf, inf, inf, inf], dtype=np.float32),
            dtype=np.float32,
        )

        # Physics and initial state configuration.
        self._delta_t = 0.01
        self._init_pos_x = 0.0
        self._init_pos_y = 0.25
        self._init_speed_x = 1.0
        self._init_speed_y = 0.0

        # Variables updated during every step.
        self._pos_x = None
        self._pos_y = None
        self._speed_x = None
        self._speed_y = None
        self._path = []
        self._step = 0

        # Rendering helpers.
        self.screen = None
        self.clock = None
        self.isopen = True

    def get_state(self):
        return np.array(
            [self._pos_x, self._pos_y, self._speed_x, self._speed_y], dtype=np.float32
        )

    def set_state(self, state):
        self._pos_x = float(state[0])
        self._pos_y = float(state[1])
        self._speed_x = float(state[2])
        self._speed_y = float(state[3])

    def _state_reset(self) -> D.T_state:
        # Reset the environment to its initial state and return the first observation.
        self._pos_x = self._init_pos_x
        self._pos_y = self._init_pos_y
        self._speed_x = self._init_speed_x
        self._speed_y = self._init_speed_y
        self._step = 0
        self._path = [(self._pos_x, self._pos_y)]
        return np.array(
            [self._pos_x, self._pos_y, self._speed_x, self._speed_y], dtype=np.float32
        )

    def _state_step(
        self, action: D.T_event
    ) -> TransitionOutcome[D.T_state, Value[D.T_value], D.T_predicate, D.T_info]:
        # Convert the raw action into a numpy array and validate its shape.
        action = np.asarray(action, dtype=np.float32).flatten()
        if action.shape != (2,):
            raise ValueError("Action must be a 2-dimensional vector [a_x, a_y].")

        # Enforce the acceleration norm constraint by applying a penalty if violated.
        action_norm = np.linalg.norm(action)
        penalty = PENALTY if action_norm < ACCELERATION_MIN else 0.0

        # Apply the action to the system dynamics.
        self._speed_x += action[0] * self._delta_t
        self._speed_y += action[1] * self._delta_t
        self._pos_x += self._speed_x * self._delta_t
        self._pos_y += self._speed_y * self._delta_t

        # Save the trajectory for rendering.
        self._path.append((self._pos_x, self._pos_y))

        self._step += 1

        # Reward is high when the object is near the center line y = 0.
        reward = exp(-fabs(self._pos_y)) + penalty

        # End the episode after a fixed horizon or if the object goes too far.
        done = (
            self._step >= HORIZON
            or fabs(self._pos_y) > 3.0
            or fabs(self._pos_x) > 20.0
        )

        obs = np.array(
            [self._pos_x, self._pos_y, self._speed_x, self._speed_y], dtype=np.float32
        )
        return TransitionOutcome(obs, Value(reward=reward), done, {})

    def _get_observation_space_(self) -> Space[D.T_observation]:
        return self.observation_space

    def _get_action_space_(self) -> Space[D.T_event]:
        return self.action_space

    def close(self):
        if self.screen is not None:
            pygame.display.quit()
            pygame.quit()
            self.isopen = False

    def _render_from(self, memory: D.T_state, **kwargs: Any) -> Any:
        # Render the trajectory in a simple 2D view, where the horizontal center line is y = 0.
        screen_width = 500
        screen_height = 100

        if self.screen is None:
            pygame.init()
            pygame.display.init()
            self.screen = pygame.display.set_mode((screen_width, screen_height))
        if self.clock is None:
            self.clock = pygame.time.Clock()

        surf = pygame.Surface((screen_width, screen_height))
        surf.fill((255, 255, 255))
        gfxdraw.hline(surf, 0, screen_width, int(screen_height / 2), (0, 0, 255))

        # Draw the visited trajectory.
        if len(self._path) > 1:
            for p in range(len(self._path) - 1):
                gfxdraw.line(
                    surf,
                    int(self._path[p][0] * 100 + 10),
                    int(screen_height / 2 + self._path[p][1] * 100),
                    int(self._path[p + 1][0] * 100 + 10),
                    int(screen_height / 2 + self._path[p + 1][1] * 100),
                    (255, 0, 0),
                )

        surf = pygame.transform.flip(surf, False, True)
        self.screen.blit(surf, (0, 0))
        return np.transpose(np.array(pygame.surfarray.pixels3d(self.screen)), axes=(1, 0, 2))
````

Now we test a trajectory of a random RL agent.

````python
from typing import *

from skdecide import *
from skdecide.builders.domain import *
from skdecide.builders.solver import *


class D(RLDomain, UnrestrictedActions, FullyObservable, Renderable):
    pass


class RandomSolver(DeterministicPolicySolver):
    T_domain = D

    def __init__(self, domain_factory, action_space) -> None:
        super().__init__(domain_factory)
        self.action_space = action_space

    def _get_next_action(self, observation: D.T_observation) -> D.T_event:
        # Return a random action sampled from the domain's valid action space.
        return self.action_space.sample()

    def _is_policy_defined_for(self, observation: D.T_observation) -> bool:
        return True
````

````python
import matplotlib.pyplot as plt
from IPython import display
%matplotlib inline

# Build a fresh domain instance and a random policy for visualization.
domain_factory = lambda: SimpleLineControlDomain()
domain = domain_factory()


def rollout(domain, solver, max_steps):
    """Simulate the domain using the solver and display each rendered frame."""
    obs = domain.reset()

    for i in range(max_steps):
        plt.imshow(domain.render())
        display.display(plt.gcf())
        display.clear_output(wait=True)
        outcome = domain.step(solver.sample_action(obs))
        obs = outcome.observation


rollout(domain=domain, solver=RandomSolver(domain_factory, domain.get_action_space()), max_steps=50)
domain.close()
````

We optimize the RL agent using RLLib.

````python
# This section configures the Ray RLLib solver that will train a policy on the domain.
# We disable CUDA and TensorFlow logging for a lightweight execution in this notebook.
````

````python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"

# Import the RL algorithm (Trainer) we would like to use.
from ray.rllib.algorithms.ppo import PPO
from skdecide.hub.solver.ray_rllib import RayRLlib
import ray
ray.init(num_gpus=0, ignore_reinit_error=True)
domain = SimpleLineControlDomain()
assert RayRLlib.check_domain(domain)
solver_factory = lambda: RayRLlib(
    domain_factory=domain_factory, algo_class=PPO, train_iterations=10
)

with solver_factory() as solver:
    # Solve domain
    solver.solve()

    # Test solution
    rollout(
        domain=domain,
        solver=solver,
        max_steps=HORIZON,
    )

domain.close()
````

### Stochastic line control

Now we make the environment stochastic by assuming that the object actuators are noised in such a way that the lateral acceleration (along the y axis) follows a Gaussian distribution centered around the lateral acceleration command with a standard deviation depending on the magnitude of the longitudinal acceleration command.
If we note $a^c_x(t)$ and $a^c_y(t)$ the acceleration command (RL actions), the actual accelerations that will act on the object are $a_x(t) = a^c_x(t)$ and $a_y(t) \sim \mathcal{N} \left( a^c_y(t), \sqrt{\vert a^c_x(t) \vert} \right)$.
Please implement this stochastic domain in a Gym environment and implement an RLLib agent that learns to solve it.

````python
class StochasticLineControlDomain(SimpleLineControlDomain):
    """Stochastic version of the line control environment.

    The longitudinal acceleration a_x is applied exactly as commanded,
    while the lateral acceleration a_y is sampled from a normal distribution:
      a_y ~ N(a_y_cmd, sqrt(|a_x_cmd|)).
    This models actuator noise that depends on the forward acceleration.
    """

    def _state_step(
        self, action: D.T_event
    ) -> TransitionOutcome[D.T_state, Value[D.T_value], D.T_predicate, D.T_info]:
        action = np.asarray(action, dtype=np.float32).flatten()
        if action.shape != (2,):
            raise ValueError("Action must be a 2-dimensional vector [a_x, a_y].")

        a_x_cmd = float(action[0])
        a_y_cmd = float(action[1])
        a_x = a_x_cmd
        noise_std = np.sqrt(abs(a_x_cmd))
        a_y = np.random.normal(loc=a_y_cmd, scale=noise_std)

        action_norm = np.linalg.norm(action)
        penalty = PENALTY if action_norm < ACCELERATION_MIN else 0.0

        self._speed_x += a_x * self._delta_t
        self._speed_y += a_y * self._delta_t
        self._pos_x += self._speed_x * self._delta_t
        self._pos_y += self._speed_y * self._delta_t

        self._path.append((self._pos_x, self._pos_y))
        self._step += 1

        reward = exp(-fabs(self._pos_y)) + penalty
        done = (
            self._step >= HORIZON
            or fabs(self._pos_y) > 3.0
            or fabs(self._pos_x) > 20.0
        )

        obs = np.array(
            [self._pos_x, self._pos_y, self._speed_x, self._speed_y], dtype=np.float32
        )
        return TransitionOutcome(obs, Value(reward=reward), done, {})


# Train a stochastic policy using Ray RLLib if available.
# This code will only execute in an environment with ray and skdecide installed.
try:
    import os
    from ray.rllib.algorithms.ppo import PPO
    from skdecide.hub.solver.ray_rllib import RayRLlib
    import ray

    os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
    os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"
    ray.init(num_gpus=0, ignore_reinit_error=True)

    stochastic_factory = lambda: StochasticLineControlDomain()
    assert RayRLlib.check_domain(stochastic_factory())
    solver_factory = lambda: RayRLlib(
        domain_factory=stochastic_factory,
        algo_class=PPO,
        train_iterations=10,
    )

    with solver_factory() as solver:
        solver.solve()

        # Test the trained policy using a rollout on the stochastic domain.
        rollout(domain=stochastic_factory(), solver=solver, max_steps=HORIZON)

    ray.shutdown()
except Exception as e:
    print("Training skipped: ", e)
````

````python
class AlternateNoisyLineControlDomain(SimpleLineControlDomain):
    """Alternative noisy model for experimentation.

    This model adds noise to the longitudinal acceleration a_x and uses a small
    constant jitter on the lateral acceleration a_y. The goal is to compare how
    different noise models affect learning.
    """

    def _state_step(
        self, action: D.T_event
    ) -> TransitionOutcome[D.T_state, Value[D.T_value], D.T_predicate, D.T_info]:
        action = np.asarray(action, dtype=np.float32).flatten()
        if action.shape != (2,):
            raise ValueError("Action must be a 2-dimensional vector [a_x, a_y].")

        a_x_cmd = float(action[0])
        a_y_cmd = float(action[1])
        a_x = np.random.normal(loc=a_x_cmd, scale=0.2)
        a_y = np.random.normal(loc=a_y_cmd, scale=0.5)

        action_norm = np.linalg.norm(action)
        penalty = PENALTY if action_norm < ACCELERATION_MIN else 0.0

        self._speed_x += a_x * self._delta_t
        self._speed_y += a_y * self._delta_t
        self._pos_x += self._speed_x * self._delta_t
        self._pos_y += self._speed_y * self._delta_t

        self._path.append((self._pos_x, self._pos_y))
        self._step += 1

        reward = exp(-fabs(self._pos_y)) + penalty
        done = (
            self._step >= HORIZON
            or fabs(self._pos_y) > 3.0
            or fabs(self._pos_x) > 20.0
        )

        obs = np.array(
            [self._pos_x, self._pos_y, self._speed_x, self._speed_y], dtype=np.float32
        )
        return TransitionOutcome(obs, Value(reward=reward), done, {})


# Example experiment with the alternate noise model.
# The rollout below helps visualize the behavior of this new noise model.
try:
    alternate_domain_factory = lambda: AlternateNoisyLineControlDomain()
    alternate_domain = alternate_domain_factory()
    rollout(domain=alternate_domain, solver=RandomSolver(alternate_domain_factory, alternate_domain.get_action_space()), max_steps=50)
    alternate_domain.close()
except Exception as e:
    print("Alternate noise experiment skipped:", e)
````
